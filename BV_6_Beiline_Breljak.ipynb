```
def myCrossCorrelation(func1, func2):
    return np.correlate(func1, func2)
```

```
input_func_1a = np.array([random.randrange(1, 255, 1) for i in range(10)])

input_func_2a = np.array([1])

output_3a = myCrossCorrelation(input_func_1a, input_func_2a)
print(output_3a)
```

```
nx = 5.5
lags = np.arange(-nx + 1, nx) # so last value is nx - 1

fig, ax = plt.subplots()
ax.plot(lags, output_3a, 'r')
ax.grid(True)
```

```
def dft_matrix(n):
    """
    Returns the Discrete Fourier Transformation Matrix of order `n`.
    """

    img = np.ones((n,n))
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            img[x,y] = (np.exp( - 2 * np.pi * 1J / n ))**(x*y)
    return img / np.sqrt(n) 

plt.subplot(121)
plt.imshow(dft_matrix(16))
plt.subplot(122)
plt.imshow(dft_matrix(32))
plt.show()
```
